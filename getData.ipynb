```python
import yfinance as yf
generation = [str(i).zfill(6) + ".SZ" for i in range(167000, 170000)]
```

```python
import pandas as pd

data = pd.DataFrame(columns=generation)

for gen in generation:
    data[gen] = yf.download(gen,'2016-01-01','2016-12-31')['Adj Close']
```

```python
data.to_excel('outputNew_10.xlsx', index=False)
```
